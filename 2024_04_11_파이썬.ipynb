```python
import os

##  단어 추가
def add_word(dictionary, word, meaning):
    dictionary[word] = meaning

## 단어 검색
def load_dictionary(filename):
    dictionary = {}
    with open(filename, 'r', encoding='utf-8') as file:
        for line in file:
            word, meaning = line.strip().split(':')
            dictionary[word] = meaning
    return dictionary


## 단어장 저장
def save_dictionary(dictionary, filename):
    with open(filename, 'w', encoding='utf-8') as file:
        for word, meaning in dictionary.items():
            file.write(f"{word}:{meaning}\n")



## 단어 표시
def disp_dictionary(filename):
    dictionary = {}
    with open(filename, 'r', encoding='utf-8') as file:
         for line in file:
            word, meaning = line.strip().split(':')
            print(f" {word} : {meaning}")
    return dictionary


##  단어장 삭제
def drop_dictionary(filename):
    try:
        os.remove(filename)
    except FileNotFoundError:
        print("단어장 파일이 존재하지 않습니다.")


def main():
    dictionary = {}

    # 단어장 파일이 있으면 불러오기
    try:
        dictionary = load_dictionary('wordlist.txt')
        print("단어장을 불러왔습니다.")
    except FileNotFoundError:
        print("단어장 파일이 존재하지 않습니다. 새로운 단어장을 생성합니다.")

    while True:
        print("\n====================")
        print("1. 1단어 추가")
        print("2. 1단어 검색")
        print("3. 1단어 삭제") ##추가
        print("4. 단어장 display")
        print("5. 단어장 저장")
        print("6. 단어장 삭제")
        print("0. 종료")
        print("====================\n")

        choice = input("선택 : ")

        if choice == '1':
            word = input("영어 단어를 입력하세요: ")
            meaning = input("뜻을 입력하세요: ")
            add_word(dictionary, word, meaning)

        elif choice == '2':
            word = input("검색할 영어 단어를 입력하세요: ")
            meaning = dictionary.get(word)
            if meaning:
                print(f"{word}: {meaning}")
            else:
                print("단어를 찾을 수 없습니다.")

        elif choice == '3':
            word = input("삭제할 영어 단어를 입력하세요: ")
            meaning = dictionary.get(word)
            print(dictionary)
            if meaning:
                print(f"{word}: {meaning}를 삭제했습니다.")
                del dictionary[word]
            else:
                print("단어를 찾을 수 없습니다.")
                continue
        elif choice == '4':
            disp_dictionary('wordlist.txt')
        elif choice == '5':
            save_dictionary(dictionary, 'wordlist.txt')
            print("단어장을 저장했습니다.")
        elif choice == '6':
            drop_dictionary('wordlist.txt')
            print("단어장을 삭제했습니다.")
            continue
        elif choice == '0':
            print("프로그램을 종료합니다.")
            break
        else:
            print("올바른 선택지를 입력하세요.")

if __name__ == "__main__":
    main()
```
