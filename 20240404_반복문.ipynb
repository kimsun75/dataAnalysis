```python
fruits=['사과','포도','딸기','키위']
for idx, fruit in   enumerate(fruits):
    print(idx, fruit)
```

```python
n=int(input('합계를 구할 숫자 입력:'))
hap=0

for i in range(1,n+1):
    if i%2 != 0:
     hap=hap+i
print(f"1~n까지의 짝수 제외한 합은 : {hap}")


```

```python
n=int(input('합계를 구할 숫자 입력:'))
hap=0
for i in range(1,n+1,2):
         hap=hap+i
print(f"1~n까지의 짝수 제외한 합은 : {hap}")
```

```python
n=int(input('합계를 구할 숫자 입력:'))
hap=sum([ i for i in range(1,n+1,2)])

print(f"1~n까지의 짝수 제외한 합은 : {hap}")
```
