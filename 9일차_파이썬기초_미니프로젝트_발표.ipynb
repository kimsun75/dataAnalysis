```python
# 주소록 프로그램 만들기

menu = ''' ---------[선택메뉴]--------
1. 연락처 입력
2. 연락처 검색
3. 연락처 삭제
4. 연락처 보기
0. 종료하기
'''

address_book = {}  # 주소록을 저장할 딕셔너리

# 함수 선언부분
def book_add():
    name = input('이름:')
    phone_number = input('전화번호:')
    email  = input('E-mail:')
    addr =input('Address:')

    # 주소록에 딕셔너리 형태로 저장 {이름 key, value 값으로 전화번호, 이메일, 주소를 딕셔너리에 저장}
    address_book[name] = {'전화번호': phone_number,'E-mail': email,'Address':addr}
    print(f"'{name}'연락처가 등록되었습니다.")

def book_search():   # 연락처 검색
    name_search = input('연락처 검색: ')
    if name_search in address_book:                 # 현재 등록된 딕셔너리에  이름(key)가 있는지 검색
        contact_info = address_book[name_search]    # 해당 이름에 대한 연락처(value) 정보 가져오기
        print(f"이름: {name_search}")                # 주소록 정보 출력
        print(f"전화번호: {contact_info['전화번호']}")
        print(f"E-mail: {contact_info['E-mail']}")
        print(f"Address: {contact_info['Address']}")
    else:
        print("연락처가 등록되어 있지 않습니다..")
        input_name = input('새로운 연락처를 등록하시겠습니니까?(등록:Y, 미등록:N):')  # 미등록된 연락처 추가할지 여부 확인
        if input_name =='Y' or 'y':
            book_add()  # 연락처 추가 함수 호출
        else:
            print(f'등록하지 않습니다.')

def book_delete():  # 연락처 삭제
    name_Delete = input('삭제할 연락처 :')     # 삭제할 연락처 이름 받기
    if name_Delete in address_book:           # 삭제할 연락처가 딕셔너리에 있는지 검색
        del address_book[name_Delete]         # 딕셔너리 주소록에서 연락처 삭제
        print(f"연락처가 삭제 되었습니다: {name_Delete}")
    else:
        print(f"등록되지 않은 이름입니다..")

def book_show():
    for name, contact_info in address_book.items():   # 주소록의 모든 연락처에 대해 반복
        print(f"이름: {name}")                        # 현재 등록된 모든 {이름 key, value 값으로 전화번호, 이메일, 주소} 출력
        print(f"전화번호: {contact_info['전화번호']}")
        print(f"E-mail: {contact_info['E-mail']}")
        print(f"Address: {contact_info['Address']}")


# 메인코드 부분
while True:
    print(menu)
    select = input('-->메뉴 선택: ')
    if select == '0':    # 프로그램 종료
        break
    elif select == '1':  # 연락처 입력
        book_add()       # 연락처 입력 함수 호출
    elif select == '2':  # 연락처 검색
        book_search()    # 연락처 검색 함수 호출
    elif select == '3':  # 연락처 삭제
        book_delete()    # 연락처 삭제 함수 호출
    elif select == '4':  # 연락처 보기
        book_show()      # 연락처 보기 함수 호출

print('프로그램을 종료합니다.')
```

```python
menu = '''
---선택메뉴---
1. 단어 등록
2. 단어 검색
3. 단어 삭제
4. 모든 단어 보기
0. 종료하기
'''

# 단어 리스트[,,,]와 딕셔너리{"키":"벨류","키":"벨류"}
vocabulary = {
    "apple": "사과",
    "banana": "바나나",
    "cat": "고양이",
    "dog": "개",
    "book": "책",
}

# 총 단어수 함수
def display_word_count():
    total_words = len(vocabulary)
    print(f"총 단어 수: {total_words}")

# 단어등록 함수
def w_add():
    enWord = input("추가할 영어단어 입력: ")
    koWord = input("추가할 한국어단어 입력: ")
    new_word = {enWord: koWord}
    vocabulary.update(new_word)
    print('단어가 추가되었습니다.')

# 단어검색 함수
def w_search():
    searchWord = input("검색할 단어를 입력: ")
    found = False # 검색결과 여부를 확인하는 변수
    for enWord, koWord in vocabulary.items():
        if searchWord in enWord or searchWord in koWord:
            print(f"영어: {enWord}, 한국어: {koWord}")
            found = True
    if not found:
        print(f"{searchWord}: 단어장에 없습니다.")

# 단어삭제 함수
def w_delete():
    deleteWord = input("삭제할 단어를 입력: ")
    for enWord, koWord in vocabulary.items():
        if deleteWord == enWord or deleteWord == koWord:
            del vocabulary[enWord]
            print(f"영어: {enWord}, 한국어: {koWord}가 삭제되었습니다.")
            return

# 프로그램 돌리기
while True:
    print(f"{menu}")
    print('-' * 30)
    select = input(f"▶ 메뉴를 선택하세요.(0~4): ")
    if select == "0":
        break
    elif select == "1":
        w_add()
    elif select == "2":
        w_search()
    elif select == "3":
        w_delete()
    elif select == "4":
        display_word_count()
        print(f"{vocabulary}")

print('프로그램을 종료합니다.')
```

```python
# 파이숑과 함께하는 UP & Down 게임
import random

desc = f'''[[ 파이숑과 함께하는 UP & Down 게임! ]] {"ㅛ"*13}
1. 파이숑이 선택할 숫자의 자릿수를 지정해주세요.
   ( 1 => 1 ~ 9 / 2 => 10 ~ 99 / 3 => 100 ~ 999 )
2. 도전 값에 따라 UP & Down 을 알려줍니다.
3. 도전 실패 5회 마다 결정적 힌트를 얻을 수 있어요.
4. 도전 시작 후 그만하기를 원하는 경우 입력창에 '0'을 넣어주세요.
5. 과연 몇번만에 파이숑의 숫자를 맞출 수 있을까요^^
{"ㅠ"*33}
'''
print(desc)

pythongNo, tryCnt = 0, 0  # 파이숑 선택 숫자, 시도 횟수

# >>>> 파이숑에게 선택할 숫자의 자릿수 받기:
digit = int(input('파이숑이 선택할 숫자의 자릿수를 입력해주세요: '))

# >>> 파이숑 숫자 선택(random)
pythongNo = random.randint(int(f'1{"0"* (digit-1) }'),int(f'{"9"* digit}'))
print(f'{"-"*66}\n>>> 파이숑 숫자 1{"0"* (digit-1) } ~ {"9"* digit} 중 선택 완료!. ({("■"*digit)})\n>>> 도전 시작~')

while True:

    tryCnt += 1

    # >>> 답변 입력
    answer = int(input(f'{"-"*50}\n파이숑이 선택한 {digit}자리의 숫자는 무엇일까요?: '))
    print('-'*66)

    # >>> 그만하기(종료)
    if answer == 0:
        tyrGo = int(input('>>> 게임을 중단하시겠습니까?(예: 1 / 아니오: 2 ): '))

        if tyrGo == 1:
            print(f'\n>>> (●´⌓`●) 파이숑이 선택한 숫자는 ( \033[91m {pythongNo} \033[30m ) 였어요.')
            # # \033[(숫자)m : 원하는 색에 대해 ANSI 표준으로 정해진 숫자 지정
            break

        else:
            continue

    # >>> 정답
    if pythongNo == answer:
        print(f'\n>>> \033[91m( {pythongNo} )정답!! ✌(❛‿❛)✌ {tryCnt}회 만에 성공했어요~')
        break

    # >>> UP & Down
    if pythongNo > answer:
        print(f'>>> \033[94m▼Down▼\033[30m 숫자를 좀 더 올려 볼까요~')
    else:
        print(f'>>> \033[95m▲UP▲\033[30m 숫자를 좀 더 내려 보아요~ ')


    # >>> 힌트
    if tryCnt % 5 == 0:
        hintDigit = int(input(f'\n>>> [⭐힌트] 몇번째 숫자를 알려드릴까요?: '))

        hint = ''
        for i in range(digit):
            if i+1 == hintDigit:
                hint += str(pythongNo)[hintDigit-1:hintDigit]
            else:
                hint += '■'

        print(f'>>> 파이숑이 선택한 숫자는 {hint} 입니다\n')
```

```python
# [미션] 영어 단어장(단어 관리) 프로그램
words = {}  # 단어 목록 딕셔너리

def w_add():
    # 단어 등록 기능
    word = input("새 단어: ")
    meaning = input(f"{word}의 뜻: ")

    if word in words:
        print(f"이미 존재하는 단어입니다: {word}")
        return

    words[word] = {"meaning": meaning,}

    print(f"단어 '{word}'가 등록되었습니다.")


def w_search():
    # 단어 검색 기능
    word = input("검색할 단어 입력: ")

    if word in words:
        print(f"\n단어: {word}")
        print(f"뜻: {words[word]['meaning']}")
    else:
        print(f"검색어 '{word}'를 찾을 수 없습니다.")


def w_delete():
    # 단어 삭제 기능
    word = input("삭제할 단어 입력: ")

    if word in words:
        del words[word]
        print(f"단어 '{word}'가 삭제되었습니다.")
    else:
        print(f"삭제할 단어 '{word}'를 찾을 수 없습니다.")


def w_show():
    # 모든 단어 보기 기능
    if not words:
        print("등록된 단어가 없습니다.")
        return

    for word, data in words.items():
        print(f"\n단어: {word}")
        print(f"뜻: {data['meaning']}")


while True:
    print("\n--- 선택메뉴 ---")
    print("1. 단어 등록")
    print("2. 단어 검색")
    print("3. 단어 삭제")
    print("4. 모든 단어 보기")
    print("0. 종료하기")

    choice = input("-----------------------선택: ")

    if choice == '1':
        w_add()
    elif choice == '2':
        w_search()
    elif choice == '3':
        w_delete()
    elif choice == '4':
        w_show()
    elif choice == '0':
        print("프로그램을 종료합니다.")
        break
    else:
        print("잘못된 입력입니다. 다시 선택해주세요.")
```

```python
print('영어 단어(단어관리) 프로그램')
menu = '''
---선택메뉴---
1. 단어 등록
2. 단어 검색
3. 단어 삭제
4. 모든 단어 보기
0. 종료하기'''
def w_add(dict):
    word = input('새단어: ')
    mean = input(f'{word}의 뜻: ')
    dict[word] = mean
def w_search(dict):
    word = input('검색할 단어: ')
    if word in dict:
        mean = dict.get(word)
        print(f'{word} : {mean}')
    else:
        print('미등록 단어')
def w_delete(dict):
    word = input('삭제할 단어: ')
    if word in dict:
        del dict[word]
        print(f"{word} 단어 삭제됨!")
    else:
        print('미등록 단어')
def w_show(dict):
    for word, mean in dict.items():
        print(f"{word}: {mean}")
def quit():
    print('프로그램을 종료합니다.')
word_dict = {}
while True:
    print(menu)
    select = input('--------------------------------------선택:')
    if select == '0':   # 0은 종료하기
        quit()
        break
    if select == '1':   # 1은 단어등록
        w_add(word_dict)
        print('--------------------------------------')
        continue
    if select == '2':   # 2은 단어검색
        w_search(word_dict)
        print('--------------------------------------')
        continue
    if select == '3':   # 3은 단어삭제
        w_delete(word_dict)
        print('--------------------------------------')
        continue
    if select == '4':   # 4은 모든 단어보기
        w_show(word_dict)
        print('--------------------------------------')
        continue
```

```python
from gtts import gTTS
from IPython.display import Audio, display
import time

def speaker(stmt):
    kor_wav = gTTS(stmt, lang='ko')
    kor_wav.save('kor.wav')
    display(Audio('kor.wav', autoplay=True))

# 파일에 내용을 추가하는 함수
def write_to_file(stmt):
    with open('output.txt', 'a') as file:  # 'a' 모드는 파일에 내용을 추가합니다.
        file.write(stmt + '\n')

#1. 단어등록
def w_add(dictionary):
    while True:
        key = input('새 단어 : ')
        if key in dictionary:
            msg = f'입력하신 {key}는 이미 등록되어 있는 단어입니다.'
            print(msg)
            speaker(msg)
            write_to_file(msg)
        else:
            break
    value = input('의 뜻 : ')
    msg = f'등록단어: {key}, 뜻: {value}'
    print(msg)
    speaker(f'새로 등록하신 단어는 {key}이고, 뜻은 {value}입니다.')
    write_to_file(msg)  # 등록된 단어를 파일에 저장
    dictionary[key] = value

#2. 단어검색
def w_search(dictionary):
    search_term = input('검색할 단어 또는 뜻 : ')
    found = False

    # key 검색
    if search_term in dictionary:
        meaning = dictionary[search_term]
        msg = f'검색하신 단어는 {search_term}이고 뜻은 {meaning}입니다.'
        print(msg)
        speaker(msg)
        write_to_file(msg)  # 검색된 단어를 파일에 저장
        found = True

    # value 검색
    value_searched = False  # value 검색 여부를 나타내는 변수 추가
    for key, value in dictionary.items():
        if value == search_term:
            msg = f'검색하신 뜻은 {value}이고 단어는 {key}입니다.'
            print(msg)
            speaker(msg)
            write_to_file(msg)  # 검색된 단어를 파일에 저장
            found = True
            value_searched = True  # value 검색이 실행된 경우 변수를 True로 설정

    # key 검색이나 value 검색이 모두 실행되지 않았고 value 검색이 실행된 적도 없는 경우에만 일부 검색 수행
    if not found and not value_searched:
        for key, value in dictionary.items():
            if search_term in key or search_term in value:
                msg = f'검색하신 일부를 포함하는 단어는 {key}이고 뜻은 {value}입니다.'
                print(msg)
                speaker(msg)
                write_to_file(msg)  # 검색된 단어를 파일에 저장
                time.sleep(5)
                found = True

    # 검색 결과가 없는 경우
    if not found:
        msg = f'"{search_term}"에 대한 검색 결과가 없습니다.'
        print(msg)
        speaker(msg)
        write_to_file(msg)  # 검색결과가 없는 단어를 파일에 저장

#3. 단어삭제
def w_delete(dictionary):
    key = input('삭제할 단어 : ')
    if key in dictionary:  # 딕셔너리에 입력된 키 값이 존재하는지 확인
        del dictionary[key]  # 키 값이 존재하면 해당 단어를 삭제
        msg = f'"{key}" 단어 삭제됨!'
        print(msg)
        speaker(f'삭제하신 단어는 {key}입니다.')
    else:
        msg = f'"{key}" 단어는 미등록 단어'
        print(msg)
        speaker(f'{key}는 미등록 단어입니다.')
    write_to_file(msg)

#4. 모든단어보기
def w_show(dictionary):
    msg = str(dictionary)
    print(msg)
    time.sleep(3)
    speaker(f'등록되어 있는 모든 단어는 {msg}입니다.')
    write_to_file(msg)

#0. 종료하기
def quit():
    msg = '프로그램을 종료합니다.'
    print(msg)
    speaker(msg)
    write_to_file(msg)

dictionary = dict()    # 빈 딕셔너리 생성

# 메뉴 출력 여부를 나타내는 변수
menu_displayed = False

while True:
    menu = '''---선택메뉴--
    1. 단어 등록
    2. 단어 검색
    3. 단어 삭제
    4. 모든 단어 보기
    0. 종료하기
    --------------------
    원하시는 메뉴를 선택해주세요'''
    print(menu, end='')

    time.sleep(5)

    if not menu_displayed:
        speaker(menu)
        menu_displayed = True

    time.sleep(2)

    select = input('선택:')

    if select == '1':
        speaker('새 단어를 등록하세요')
        w_add(dictionary)

    elif select == '2':
        speaker('단어를 검색하세요')
        w_search(dictionary)

    elif select == '3':
        speaker('삭제할 단어를 입력하세요')
        w_delete(dictionary)

    elif select == '4':
        speaker('모든 단어를 보여주세요')
        w_show(dictionary)

    elif select == '0':
        quit()
        break
```

```python
# 생일을 저장할 딕셔너리
anniversary_dict = {}

# 함수 정의

# 1. 생일 등록
def a_add():
    day = input("생일자 이름 혹은 호칭을 입력하세요.: ")
    meaning = input("생일 날짜를 입력하세요.: ")
    anniversary_dict[day] = meaning
    print("생일이 성공적으로 등록되었습니다.")

# 2. 생일 검색
def a_search():
    day = input("누구 생일이 궁금해요?: ")
    if day in anniversary_dict:
        print(f"{day}: {anniversary_dict[day]}")
    else:
        print("미등록 생일")

# 3. 생일 삭제
def a_delete():
    day = input("삭제할 생일자 이름 혹은 호칭을 입력하세요: ")
    if day in anniversary_dict:
        del anniversary_dict[day]
        print("생일이 성공적으로 삭제되었습니다.")
    else:
        print("미등록 기념일")

# 4. 모든 생일 보기
def a_show():
    if anniversary_dict:
        print("===== 생일 목록 =====")
        for day, meaning in anniversary_dict.items():
            print(f"{day}: {meaning}")
        print("=====================")
    else:
        print("등록된 생일이 없습니다.")

# 0. 종료하기
def quit():
    print("프로그램을 종료합니다.")
    exit()

# 메뉴 출력
def print_menu():
    print("\n--- 선택 메뉴 ---")
    print("1. 생일 등록")
    print("2. 생일 검색")
    print("3. 생일 삭제")
    print("4. 모든 생일 보기")
    print("0. 종료하기")

# 메인 함수
def main():
    while True:
        print_menu()
        choice = input("메뉴 번호를 입력하세요: ")

        if choice == '1':
            a_add()
        elif choice == '2':
            a_search()
        elif choice == '3':
            a_delete()
        elif choice == '4':
            a_show()
        elif choice == '0':
            quit()
        else:
            print("올바른 메뉴 번호를 입력하세요.")

# 프로그램 실행
if __name__ == "__main__":
    main()
```
